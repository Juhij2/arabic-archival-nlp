```python
import json
import re
import os

# Load our OCR results with correct path
with open("data/ocr_output/all_ocr_results.json", 
          encoding="utf-8") as f:
    ocr_results = json.load(f)

print("✓ Loaded OCR results")
for name, data in ocr_results.items():
    print(f"  {name}: {data['char_count']} characters")
```

```python
import re

def clean_arabic_ocr(text):
    # Normalize Alef forms
    text = re.sub('[إأآا]', 'ا', text)
    
    # Normalize Yeh forms
    text = re.sub('[يى]', 'ي', text)
    
    # Remove non-Arabic noise
    arabic_pattern = re.compile(
        r'[^\u0600-\u06FF\u0750-\u077F'
        r'\uFB50-\uFDFF\uFE70-\uFEFF\s\.\،\,\:\!\?]'
    )
    text = arabic_pattern.sub(' ', text)
    
    # Fix broken article
    text = text.replace('ا ل', 'ال')
    
    # Clean whitespace
    text = re.sub(r' +', ' ', text)
    text = re.sub(r'\n+', '\n', text)
    
    # Remove very short lines
    lines = text.split('\n')
    lines = [l.strip() for l in lines if len(l.strip()) > 3]
    
    return '\n'.join(lines).strip()

print("✓ Cleaning function ready")
```

```python
os.makedirs("data/cleaned_text", exist_ok=True)

cleaned_results = {}

for img_name, data in ocr_results.items():
    print(f"\n{'='*50}")
    print(f"Cleaning: {img_name}")
    
    raw_text = data['text']
    cleaned = clean_arabic_ocr(raw_text)
    
    cleaned_results[img_name] = {
        "description": data['description'],
        "raw_text": raw_text,
        "cleaned_text": cleaned,
        "raw_chars": len(raw_text),
        "clean_chars": len(cleaned)
    }
    
    print(f"Raw chars:     {len(raw_text)}")
    print(f"Cleaned chars: {len(cleaned)}")
    print(f"\nCLEANED TEXT:")
    print(cleaned)
    
    # Save individual file
    txt_name = img_name.replace('.jpg', '_clean.txt')
    with open(f"data/cleaned_text/{txt_name}",
              "w", encoding="utf-8") as f:
        f.write(cleaned)
    print(f"\n✓ Saved!")

# Save combined JSON
with open("data/cleaned_text/all_cleaned.json",
          "w", encoding="utf-8") as f:
    json.dump(cleaned_results, f, 
              ensure_ascii=False, indent=2)

print(f"\n✓ All documents cleaned and saved!")
```
